```python
#!pip install sentence-transformers
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


class Similarity:

  def __init__(self):
    self.transformer = SentenceTransformer('/content/content/sentence-transformers_paraphrase-xlm-r-multilingual-v1')

  def calculate(self, query, corpus):
    query_vec = self.transformer.encode(query).reshape(1,-1)
    corpus_vec = self.transformer.encode(corpus)
    self.sim_score = cosine_similarity(query_vec, corpus_vec)
    
    return self.sim_score
```

```python
check = Similarity()
```

```python
query = 'ular makhluk ovovivipar'

s1 = 'ular berkembang biak secara ovovivipar'
s2 = 'ular bertelur lalu melahirkan'

corpus = [s1, s2]
```

```python
check.calculate(query, corpus)
```
